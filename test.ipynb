```python
import numpy as np

def calculate_angle(a, b, c):
    a = np.array(a)  # 첫 번째 점
    b = np.array(b)  # 중간 점 (관절)
    c = np.array(c)  # 세 번째 점

    # 각 축에 대한 벡터 계산
    ba = a - b  # b에서 a로의 벡터
    bc = c - b  # b에서 c로의 벡터

    # 각 축에 대한 각도 계산
    cos_angle_x = np.dot(ba[[1, 2]], bc[[1, 2]]) / (np.linalg.norm(ba[[1, 2]]) * np.linalg.norm(bc[[1, 2]]))
    cos_angle_y = np.dot(ba[[0, 2]], bc[[0, 2]]) / (np.linalg.norm(ba[[0, 2]]) * np.linalg.norm(bc[[0, 2]]))
    cos_angle_z = np.dot(ba[[0, 1]], bc[[0, 1]]) / (np.linalg.norm(ba[[0, 1]]) * np.linalg.norm(bc[[0, 1]]))

    angle_x = np.arccos(cos_angle_x) * 180.0 / np.pi
    angle_y = np.arccos(cos_angle_y) * 180.0 / np.pi
    angle_z = np.arccos(cos_angle_z) * 180.0 / np.pi

    # 결과 반환
    return (round(angle_x, 4), round(angle_y, 4), round(angle_z, 4))  # 각 축에 대해 소수점 네 자리까지 반올림
```

```python
aa = AS[['RightLeg', 'RightUpLeg', 'Hips']]
alist = []

pre_angle_x = 0
pre_angle_y = 0
pre_angle_z = 0 

for i in range(len(aa)):
    # Convert tuple strings to actual tuples if necessary
    a = eval(aa['RightLeg'][i]) if isinstance(aa['RightLeg'][i], str) else aa['RightLeg'][i]
    b = eval(aa['RightUpLeg'][i]) if isinstance(aa['RightUpLeg'][i], str) else aa['RightUpLeg'][i]
    c = eval(aa['Hips'][i]) if isinstance(aa['Hips'][i], str) else aa['Hips'][i]

    angle_x ,angle_y, angle_z  = calculate_angle(a, b, c)
    
    angle_x = pre_angle_x if angle_x == 0 or np.isnan(angle_x) else angle_x
    angle_y = pre_angle_y if angle_y == 0 or np.isnan(angle_x) else angle_y 
    angle_z = pre_angle_z if angle_z == 0 or np.isnan(angle_x) else angle_z 

    alist.append((angle_x ,angle_y, angle_z))
    
    pre_angle_x = angle_x  
    pre_angle_y = angle_y      
    pre_angle_z = angle_z  

alist
```
